```python
def draw_square(image:Image, top_x, top_y, bottom_x, bottom_y, fill = 'red', width = 2):
    draft = image.copy
    draw.line((top_x, top_y, bottom_x, bottom_y), fill = fill, width = width)
    draw.line((top_x, top_y, top_x, bottom_y), fill = fill, width = width)
    draw.line((top_x, bottom_y, bottom_x, bottom_y), fill = fill, width = width)
    draw.line((bottom_x, top_y, bottom_x, bottom_y), fill = fill, width = width)
    draw = ImageDraw.Draw(draft)
    return draft
```
